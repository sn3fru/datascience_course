Uma Regra: O B tem que ser o **contrafactual** do A. E o que é um contrafactual?

Wikipedia: Em metafísica e em lógica modal contrafatual é a situação ou evento que não aconteceu, mas poderia ter acontecido. (A situação ou evento que aconteceu é chamada de atual.) Se diz que o evento contrafatual faz parte de um mundo possível, enquanto o evento atual faz parte do mundo real.
